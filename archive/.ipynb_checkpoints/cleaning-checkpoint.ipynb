# Dealing with Indeed's search ranking
We don't know how Indeed ranks job search results, and it's likely that the method isn't random. We've tried to sample from as far across the pages of search results as possible, but generally there's still a limitation with that, given that we didn't sample all of the jobs available via search. Moreover there's the question of whether the job posts we're seeing on Indeed's page for the day of scraping are representative of Q4 2020 anyway. To get around this, I will be randomly sampling from each dataset for a fraction of the job posts. 


```python
len(clean_DS)
```

```python
sampled_clean_DS = clean_DS.sample(frac=0.25, axis=0, random_state = 12345)
sampled_clean_ML = clean_ML.sample(frac=0.25, axis=0, random_state = 12345)
```
